```python
# 8. What was the count of matches played in each season?
def year_extractor(date):
    return int(date[:4])

year = [year_extractor(date) for date in matches["date"]]
matches["year"] = year

matches.groupby("year").count()['id']
```

```python
matches.drop(columns=['year'], inplace=True)
```

```python
# 9. How many runs were scored in each season?
merged = pd.merge(ball_by_ball, matches, on='id')
merged.groupby('year')['total_runs'].sum()
```

```python
# 10. Find the top 10 teams which has won the most number of matches
matches.groupby('winner').count().sort_values('result', ascending=False).head(10)['result']
```

```python
# 11. Find the Top 10 Players With Most Player of Match

matches.groupby('player_of_match').count().nlargest(10, 'result')['result']
```

```python
# 12. Find the Percentage of choosing field after winning the toss
total_matches = matches.count()['id'] # 816
total_toss_field = matches.groupby('toss_decision').count()['id']['field'] # 496
percentage = total_toss_field * 100 / total_matches
print(f"Percentage of matches where field was chosen as a toss_decision: {round(percentage, 2)}%")
```

```python
# 13. Find the Percentage of choosing to bat first after winning the toss
total_matches = matches.count()['id'] # 816
total_toss_bat = matches.groupby('toss_decision').count()['id']['bat'] 
percentage = total_toss_bat * 100 / total_matches
print(f"Percentage of matches where bat was chosen as a toss_decision: {round(percentage, 2)}%")
```

```python
# 14. Find the ratio of Bat to Field count
total_toss = matches.groupby('toss_decision').count()['id']
bat = total_toss['bat']
field = total_toss['field']
print(f'Bat by Field Ratio: {round(bat / field, 2)}')
```

```python
# 15. Find the count of Loss vs Wins vs Ties

print('----------Win List------------')
winnerList = matches['winner'].value_counts()
print(winnerList)


lossList = {'Royal Challengers Bangalore': 0, 'Mumbai Indians': 0, 'Kolkata Knight Riders': 0, 'Chennai Super Kings': 0, 'Kings XI Punjab': 0,
    'Rajasthan Royals': 0, "Delhi Daredevils": 0, 'Sunrisers Hyderabad': 0, 'Deccan Chargers':0,'Delhi Capitals':0, 'Gujarat Lions':0,             
     'Pune Warriors':0, 'Rising Pune Supergiant': 0,'Kochi Tuskers Kerala':0, 'Rising Pune Supergiants':0}
tieList = {'Royal Challengers Bangalore': 0, 'Mumbai Indians': 0, 'Kolkata Knight Riders': 0, 'Chennai Super Kings': 0, 'Kings XI Punjab': 0,
    'Rajasthan Royals': 0, "Delhi Daredevils": 0, 'Sunrisers Hyderabad': 0, 'Deccan Chargers':0,'Delhi Capitals':0, 'Gujarat Lions':0,             
     'Pune Warriors':0, 'Rising Pune Supergiant': 0,'Kochi Tuskers Kerala':0, 'Rising Pune Supergiants':0}
def cal_loss_and_draw():
    for i in range(matches.shape[0]):
        t1, t2, w = matches['team1'][i], matches['team2'][i], matches['winner'][i]
        if t1 == w:
            lossList[t2] += 1
        elif t2 == w:
            lossList[t1] += 1
        else:
            tieList[t1] += 1
            tieList[t2] += 1
cal_loss_and_draw()
print('----------Loss List------------')
print(lossList)
print('----------Tie List-------------')
print(tieList)
```
